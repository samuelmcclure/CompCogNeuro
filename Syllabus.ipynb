## Course Objectives

This course will guide you through several of the foundational models in computational neuroscience that inform theories about how the brain structures cognition. Current research in computational cognitive neuroscience builds from or is informed by these models. The goal of this course is to introduce you to these core models to permit you to understand and explore more advanced topics.

There are different ways the understand the models that we will discuss this semester. One approach is to focus on cognition, aiming to understand the models conceptually with a rough idea for how they work. Alternatively, you can focus on the mathematics that underlies the model dynamics. This will let you understand why the models work and what their limits are. We will take a third approach: we will build the models and explore how they work through experimentation.

In order to build a model, it is imperative that you know some computer programming. This class will guide you through the basics of the programming language Python. The goal will not be to become a computer scientist capable of writing professional code. Instead, you will learn <b>scientific programming</b>, writing code that works and that is understandable even if it is not the most efficient or extendable.

After learning the fundamental of Python, the course will be divided into four sections. We will study the Hodgkin-Huxley model of voltage-gated channels and action potentials. We will then investigate simple neuron netwoks that adjust synaptic weights to learn associations between inputs. Cognition is shaped by motivational influences, and the third section will investigate how reward can direct learning within neural networks. Finally, the final section will investigate how perception may arise from inference over populations of neurons.

As we proceed through the course, you will need to complete weekly homework assignments. The end of each section will be capped by a longer assignment.

### Working in Groups

This class does not assume that you have prior experience with computer programming. We will cover basics for how to define and manipulate variables, write functions, and generate plots. However, the only way for you to learn how to program is through trial and error. Many example will be provided for you to learn from, and homeworks will encourage you to combine operations in new ways. But, you will still find yourself dealing with frustration and you try to make things work. Trial and error often proceeds much more rapidly when you work with other people. You are encouraged to attend office hours or email the instructors when you are stuck. Even better is to work in a group with other members of the class, so that you can teach eachother.

One problem with working in groups is that it creates the opportunity to complete work without full comprehension. This material in this class builds progressively over the semester. If you do not understand something early on then it may be problematic later. <b>You are therefore encouraged to work in groups, but the work you submit should be your own.</b> You should be able to explain how your solutions work and reproduce the assignments on your own. As long as you accomplish this then completing class assignments in groups is encouraged.

### ChatGPT

AI tools have changed the way people write code. For example, [ChatGPT](https://chat.openai.com) can produce Python code from simple English prompts. I asked ChatGPT to `write python code to plot 10 cycles of a sine wave`. It output the code below. Running the code generates exactly the plot I was hoping for.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 20*pi with a step of 0.01
x = np.arange(0, 20 * np.pi, 0.01)

# Generate y values by calculating the sine of x
y = np.sin(x)

# Plot the sine wave
plt.plot(x, y)

# Set the x-axis label
plt.xlabel('x')

# Set the y-axis label
plt.ylabel('y')

# Set the title of the plot
plt.title('10 Cycles of a Sine Wave')

# Display the grid
plt.grid(True)

# Display 10 cycles of the sine wave
plt.xlim(0, 20 * np.pi)

# Show the plot
plt.show()
```

You should use AI tools for this course. There will probably be instances in which you will be able to solve homework problems by querying ChatGPT. However, there are two problems with doing this. First, if you rely on solutions produced by AI that you do not understand, then you may create problems with understanding later in the semester. Second, AI code is not perfect. I have used ChatGPT to create code that does not actually work. This is a great starting point. You can usually read the code, find the error and fix it. However, making these corrections requires that you understand how the code works!
